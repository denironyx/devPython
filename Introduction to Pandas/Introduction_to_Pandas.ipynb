- Generates descriptive statistics that summarize the central tendency, dispersion and shape of a datasetâ€™s distribution, excluding NaN values. <br>

- To have the summary statistics for categorical variables, we should add as a parameter 'include = ['O']'. This will give us the count, the quantity of unique values, the most frequent value and the frequency for the most frequent value.<br>

##### What is the frequency, the minimum, the maximum and the mean value of year of birth of our customers that have an recomendation level greater than 4?

**Step 21** - Using the method `agg()`, obtain the count, the min, the max and the mean for the variable `Year_Birth` for the onservations where the Recomendation is greater or equal than 4.

### What is Seaborn?
Seaborn is a graphical package that provides a high-level interface to Matplotlib, in an easy way. <br>

Seaborn allows to:<br>
- Using default themes that are aesthetically pleasing;<br>
- Setting custom colour pallets;<br>
- Making attractive statistical plots;<br>
- Easily and flexibly displaying distributions;<br>
- Visualizing information from matrices and DataFrames.<br>

Seaborn is a complement and not a substitute for Matplotlib.<br>

One of the greatest strengths of seaborn is its diversity of plotting functions. For example, to make a scatter plot, we just need to write one line of code, using the function lmplot().<br>
